Problem 1: Mobile Recharge Validation System

```python

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        return False, "Amount must be at least ₹50."
    elif amount not in valid_plans:
        return False, f"Invalid plan. Please choose from {valid_plans}."
    return True, "Recharge Successful!"

while True:
    amt = int(input("Enter recharge amount: "))
    success, message = validate_recharge(amt)
    print(message)
    if success: break
    print("Please try again.\n")
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock_dict):
    for product, qty in stock_dict.items():
        if qty < 15:
            print(f"ALERT: {product} is low on stock ({qty} left)!")
        else:
            print(f"Status: {product} stock is OK.")

inventory = {"Apples": 20, "Bananas": 10, "Oranges": 14, "Mangoes": 50}
check_inventory(inventory)
```

Problem 3: Student Result Processing System

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark
    avg = total / len(marks)
    return "Pass" if avg >= 50 else "Fail", avg

student_marks = [55, 42, 68, 70, 30]
status, average = process_result(student_marks)
print(f"Average: {average} | Result: {status}")
```

Problem 4: Cab Fare Estimator

```python
def calculate_fare(km, is_peak):
    fare = 50 + (km * 12)
    if is_peak:
        fare *= 1.25  # 25% extra
    return fare

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower() == 'yes'
    
    total_fare = calculate_fare(distance, peak)
    print(f"Estimated Fare: ₹{total_fare}")
    
    if input("Calculate another? (y/n): ") != 'y': break
```

Problem 5: Employee Attendance Checker

```python
def check_eligibility(attendance_list):
    present_days = 0
    for status in attendance_list:
        if status == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    return "Eligible" if percentage >= 75 else "Not Eligible", percentage

attendance = ["P", "P", "A", "P", "P", "A", "P"]
status, pct = check_eligibility(attendance)
print(f"Attendance: {pct:.2f}% | Status: {status}")
```

Problem 6: Password Strength Checker

```python
def check_password(password):
    if len(password) < 8:
        return "Weak: Too short."
    
    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
            
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak: Missing digit or special character (@#$)."

pwd = input("Set your password: ")
print(check_password(pwd))
```
