```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find all employees with highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

# Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

import string
from collections import Counter

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Split into words
words = query.split()

# Count frequency
word_count = Counter(words)

# Filter words appearing more than once
repeated_words = {word: count for word, count in word_count.items() if count > 1}

print(repeated_words)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Store valid readings (even numbers) with index
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

from collections import Counter

# Extract domains
domains = [email.split('@')[1] for email in emails]

# Count domains
domain_count = Counter(domains)

total_emails = len(emails)

# Calculate percentage
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
average_sales = sum(sales) / len(sales)

# Threshold = 30% above average
threshold = average_sales * 1.3

# Detect spikes
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Problem 6: Duplicate User ID Detection

from collections import Counter

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count IDs
id_count = Counter(user_ids)

# Display duplicates only
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
