Problem 1: Unique Words in a Sentence

```python
# Input sentence
sentence = input("Enter sentence: ")

# Split the sentence into a list of words (default delimiter is space)
words = sentence.lower().split()

# Convert the list to a set to extract unique elements
unique_words = set(words)

# Display results
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find the employee name with the maximum salary value
highest_paid_employee = max(employees, key=employees.get)
salary = employees[highest_paid_employee]

print(f"Highest Salary: {highest_paid_employee} - {salary}")
```

Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with the first element
maximum = numbers[0]
minimum = numbers[0]

# Iterate to compare values
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```

Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
threshold = 1000
count = 0

# Loop through prices and increment count if conditions are met
for price in prices:
    if price > threshold:
        count += 1

print(f"Products above {threshold}: {count}")
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

# Count occurrences of "P"
present_days = attendance.count("P")
total_days = len(attendance)

# Calculate percentage
percentage = (present_days / total_days) * 100

print(f"Attendance Percentage: {percentage}")
```

Problem 6: Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

# Sets do not allow duplicate entries
unique_numbers = set(phone_numbers)

print(f"Unique phone numbers: {unique_numbers}")
```

Problem 7: Count Character Frequency

```python
input_string = "pythonp"
frequency = {}

# Iterate through each character
for char in input_string:
    # If char is already in dict, increment; else, set to 1
    frequency[char] = frequency.get(char, 0) + 1

print(frequency)
```
